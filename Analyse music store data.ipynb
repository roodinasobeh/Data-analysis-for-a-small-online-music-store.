# ​Task 3: Data Integration & Aggregation

# ​Which country has the most customers?

```python
df_customer.head()
```

```python
customers_grouped_by_country = df_customer.groupby(by='country').size()
customers_grouped_by_country
```

```python
customers_grouped_by_country = df_customer.groupby(by='country').size() # الاسهل نعمل sorted ليهم وناخد اندكس صفر علطول اسهل
most_country = customers_grouped_by_country.idxmax() 
country_value = customers_grouped_by_country.max()
print('The country with most customers is', most_country, 'with' ,country_value,' customers')
```
